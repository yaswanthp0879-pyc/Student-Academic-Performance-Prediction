### Question 1: Data Loading (2 Marks)
Download `student_performance_dataset.csv` dataset from https://www.kaggle.com/datasets/aiexplorer77/academic-performance-prediction. Import pandas and load the dataset. Print the first 5 rows of the DataFrame.

**Expected Output:**
```
A pandas DataFrame displaying the first 5 rows of the dataset.
```

````python
import pandas as pd

df = pd.read_csv('student_performance_dataset.csv')

print(df.head())
````

### Question 2: Identify Column Types (2 Marks)
Identify and print the list of numerical and categorical column names in the dataset.

**Expected Output:**
```
Lists of numerical and categorical column names.
```

````python
numerical_cols = df.select_dtypes(include=['int64','float64']).columns.tolist()

categorical_cols = df.select_dtypes(include=['object']).columns.tolist()

print("Numerical Columns:")
print(numerical_cols)

print("\nCategorical Columns:")
print(categorical_cols)
````

### Question 3: Missing Values (2 Marks)
Check for missing values in the dataset and print the sum of missing values for each column.

**Expected Output:**
```


````python
print(df.isnull().sum())
````

### Question 4: Data Imputation (3 Marks)
Fill missing numerical values with the median and missing categorical values with the mode. Print the total number of missing values remaining in the entire dataset.



````python
for col in numerical_cols:
    df[col] = df[col].fillna(df[col].median())

for col in categorical_cols:
    df[col] = df[col].fillna(df[col].mode()[0])

print("Remaining Missing Values:")
print(df.isnull().sum().sum())
````

### Question 5: Label Encoding (3 Marks)
Use `LabelEncoder` from `sklearn.preprocessing` to encode all categorical columns into numerical values. Print the first 5 rows of the updated DataFrame.

**Hint:** Loop through the categorical columns and apply `.fit_transform()`.

**Expected Output:**
```
A pandas DataFrame head with all columns containing numerical values.
```

````python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

for col in categorical_cols:
    df[col] = le.fit_transform(df[col])

print(df.head())
````

### Question 6: Feature and Target Separation (3 Marks)
Separate the features `X` and the target `y` (assuming 'performance_category' is the target). Print the shapes of `X` and `y`.

**Expected Output:**
```
Shapes of X and y (e.g., (n_samples, n_features) and (n_samples,)).
```

````python
X = df.drop('performance_category', axis=1)

y = df['performance_category']

print("Shape of X:", X.shape)
print("Shape of y:", y.shape)
````

### Question 7: Train-Test Split (3 Marks)
Split the data into training and testing sets with a test size of 20% and `random_state=42`. Print the shapes of `X_train` and `X_test`.

**Expected Output:**
```
Shapes of X_train and X_test.
```

````python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42
)

print("X_train shape:", X_train.shape)
print("X_test shape:", X_test.shape)
````

### Question 8: Feature Scaling (3 Marks)
Scale the features using `MinMaxScaler` from `sklearn.preprocessing`. Fit on `X_train` and transform both `X_train` and `X_test`. Print the minimum and maximum values of the scaled `X_train`.



````python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

X_train = scaler.fit_transform(X_train)

X_test = scaler.transform(X_test)

print("Minimum value:")
print(X_train.min())

print("Maximum value:")
print(X_train.max())
````

### Question 9: Decision Tree Classifier (5 Marks)
Initialize and train a `DecisionTreeClassifier` with `random_state=42`. Make predictions on the test set and print the accuracy score.

**Expected Output:**
```
A numerical value representing the accuracy score.
```

````python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

model = DecisionTreeClassifier(random_state=42)

model.fit(X_train, y_train)

y_pred = model.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)

print("Accuracy:", accuracy)
````

### Question 10: Feature Importance (4 Marks)
Extract the feature importances from the trained Decision Tree model. Print a dictionary or a pandas Series mapping feature names to their importance scores.

**Hint:** Access `.feature_importances_` on the trained model and combine it with `X.columns`.

**Expected Output:**
```
A mapping of feature names to their importance scores.
```

````python
importance = pd.Series(
    model.feature_importances_,
    index=X.columns
)

print(importance.sort_values(ascending=False))
````
